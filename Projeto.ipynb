# Importar pandera para tratar dados do tipo DATA

```python
import pandera as pa
```

# Realizar o carregamento do arquivo em formato csv para tratar os dados.
### sep para mudar o separador de ',' para ';'
### Parse_dates para mudar o tipo de dados de Objeto para Data
### dayfist para mudar o formato do compo YYYY/MM/DD

```python
df = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
df.head(10)
```

```python
df.tail(10)
```

## Gerar esquema com pandera para validar tipo de dados

```python
schema = pa.DataFrameSchema(
    columns = {
        #"codigo" : pa.Column(pa.Int, required=False), #Caso tenha um campo que veja se dados e a gente queira permitir
        "codigo_ocorrencia" : pa.Column(pa.Int),
        "codigo_ocorrencia2" : pa.Column(pa.Int),
        "ocorrencia_classificacao" : pa.Column(pa.String),
        "ocorrencia_cidade" : pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2)),
        "ocorrencia_aerodromo" : pa.Column(pa.String),
        "ocorrencia_dia" : pa.Column(pa.DateTime),
        "ocorrencia_hora" : pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$') ,nullable=True), #nullable para permitir campo nulos
        "total_recomendacoes" : pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```
